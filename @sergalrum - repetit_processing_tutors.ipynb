## Описание данных

Сервис передает контакты клиента (ученика) репетитору. Если репетитор начинает заниматься с учеником, то он должен платить сервису комиссию от каждого занятия.

Но в реальности так происходит не всегда. Иногда, это из-за того, что репетитор звонит по телефону и ему просто не отвечают. Некоторые репетиторы плохо договариваются о занятиях или обманывают.

Сервис теряет деньги каждый раз, когда отдаёт заявку неэффективному репетитору.

---

Заказчику нужно как можно раньше понять, что репетитор недобросовестный или мошенник, чтобы отключить его от сервиса и отдавать заявки ответственным репетиторам.

```python
def def_corr_table(df, MAX_UNIQUE):
    """Функция получения списка интервальных признаков и матрицы корреляций"""
    
    INTERVAL_COLS = []
    
    for col in list(df.columns):
        if len(df[col].unique())>MAX_UNIQUE:
            INTERVAL_COLS.append(col)
    
    print("Список признаков в INTERVAL_COLS:")
    display(INTERVAL_COLS)
    table = df.phik_matrix(interval_cols=INTERVAL_COLS)
    return INTERVAL_COLS, table
```

```python
def def_date_change(DF, COL, DISPLAY, CHANGE):
    """Функция для конвертации дат в количество дней до выбранной текущей даты """
    
    local = (today - pd.to_datetime(DF[COL], errors='coerce')).dt.days
    
    if DISPLAY:
        print(f"Измененные значения по столбцу {COL}:")
        display(local.head())
        print(f'Вывод с изменением датафрейма по столбцу {COL}') if CHANGE else print(f'Вывод без изменений датафрейма по столбцу {COL}')
    
    if CHANGE:
        DF.loc[:, COL] = local
    
    print('_' * 100)
    
    del local
```

```python
def proba_plot(data, column):
    """Функция для выведения распределения уникальных значений по целевому признаку + построение barplot"""
    array0 = []
    array1 = []
    unique_list = list(sorted(data[column].unique()))
    for i in range(len(unique_list)):
        all_unique = data.loc[data[column] == unique_list[i], column].count()

        target_1 = data.loc[(data[column] == unique_list[i]) & (data['blocked'] == 1), column].count()
        proba1 = np.round((target_1 / all_unique), 2)
        target_0 = data.loc[(data[column] == unique_list[i]) & (data['blocked'] == 0), column].count()
        proba0 = np.round((target_0 / all_unique), 2)
        
        array1.append([unique_list[i], proba1, 'Есть подозрения'])
        array0.append([unique_list[i], proba0, 'Без подозрений'])

    array = np.concatenate([array0, array1], axis=0)
    df = pd.DataFrame(data=array, columns = ['Уникальные значения столбца '+column, 'Доля репетиторов', 'Hue'])
    df['Доля репетиторов'] = df['Доля репетиторов'].astype('float')
#     df = df.sort_values(['Уникальные значения столбца '+column], key=pd.to_numeric)
    display(df)
    
    sns.barplot(data = df, x = 'Уникальные значения столбца '+column, y = 'Доля репетиторов', hue = 'Hue')
    plt.show()
    
    print()
    print('Корреляция Phik:', np.round(corr_table.loc[column]['blocked'], 2))
    print('-----------------------------------------------------------------------------------------------')
    print()
    
    return None
```

### Предложенные синтетические признаки

#### Несоответствие цены в заявке, в анкете репетитора и реальной

Цена в заявке - df_orders.lesson_price  
Цена в анкете - df.lesson_cost  
Реальная цена - df.mean_lesson_price

Посчитаем разницу между выбранными значениями и получим количество заявок из имеющейся информации.

```python
def def_same_price(row):
    """Функция для определения разницу между выбранными значениями и получение количества заявок"""
    
    teacher_id = row['id']
    
    application_price = df_orders.query('working_teacher_id==@teacher_id')['lesson_price'] # цены в заявках
    profile_price = row['lesson_cost'] # цена в анкете
    real_price = row['mean_lesson_price'] # реальная средняя цена
    
    sum_mean_diff = np.sum(np.diff(np.sort([np.mean(application_price), profile_price, real_price])))
    count_orders = len(application_price)
    
#     print(count_unique_order_price)
    return [sum_mean_diff, count_orders]
# df.apply(def_same_price, axis=1)
```

```python
%%time
df[['sum_mean_diff', 'count_orders']] = df.apply(lambda row: pd.Series(def_same_price(row)), axis=1)
```

```python
# Функция отображения важности признаков модели
def plot_feature_importance(feature_names, feature_scores, top_feats=20, title='Importance',
                            x_label='Importance', y_label='Features'):
    tuples = [(name, round(score, 3)) for name, score in zip(feature_names, feature_scores)]
    tuples = sorted(tuples, key=lambda x: x[1])[-top_feats:]
    
    labels, values = zip(*tuples)
    _, ax = plt.subplots(1, 1, figsize=(10, 8))
    ylocs = np.arange(len(values))
    
    ax.barh(ylocs, values, align='center', height=0.4)
    for x, y in zip(values, ylocs):
        ax.text(x + 1, y, x, va='center')
        
    ax.set_yticks(ylocs)
    ax.set_yticklabels(labels)
    xlim = (0, max(values) * 1.1)
    ax.set_xlim(xlim)
    ylim = (-1, len(values))
    ax.set_ylim(ylim)
    
    ax.set_title(title)
    ax.set_xlabel(x_label)
    ax.set_ylabel(y_label)
    ax.grid(True)
```

```python
importances = cbc_gs.best_estimator_.feature_importances_
plot_feature_importance(X_train, importances, top_feats=X_train.shape[1], title='CatBoost feature importance', 
                        x_label='Importance')
```

Наиболее важными признаками для модели стали `Процент начала занятий`, `Количество занятий на ученика` и `Количество заявок у репетитора`.

# Рекомендации для бизнеса

При оценке распределений выявлена закономерности большей добросовестности репетиторов, подтвердивших телефон и почту (однако для обучения не удалось использовать, поскольку слишком малый процент репетиторов ферефицируют их).

В качестве упрощения отслеживания добросовестных репетиторов имеет смысл стимулирование верификации данных.
